```python
from sklearn import svm
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
X = [[16.1, 26], [46.5, 25], [72.3, 25.2], [82.6, 25], [85.2, 25.5], [85.7, 26.2], [81.6, 26.5]]
y = [0, 2.5, 5, 7.5, 10, 12.5, 15]
df = pd.DataFrame(X,columns=["humidity","temperature"])
df_ = pd.DataFrame(y,columns=["water"])
print(df)
print(df_)
X = df.iloc[:,1:2].values
y = df.iloc[:,1].values
```

```python
regr_soil = svm.SVR()
regr_soil.fit(X, y)
```

```python
index_minX = np.where(X == np.amin(X))
index_maxX = np.where(X == np.amax(X))
print(index_minX)
print(index_maxX)
X_grid = np.arange(min(X),max(X),0.01)
X_grid = X_grid.reshape(len(X_grid),1)
plt.scatter(X,y,color="blue")
plt.plot(X_grid, regr_soil.predict(X_grid), color="red")
plt.title("water needed")
plt.xlabel("{humidity,temperature}")
plt.ylabel("water")
ax = plt.gca()
ax.get_xaxis().set_visible(False)
plt.show()
        
```
